```python
from numpy import NaN
import numpy as np
import pandas as pd

map = pd.read_csv('original-files/english_hindi_id_mapping.tsv', names=["hindi_id", "hindi_category", "english_id", "english_category", "type_link"], sep='\t', header=None, keep_default_na=False)

english = pd.read_csv('original-files/english_synset_data.tsv', names=["english_id", "english_category", "english_synset_words", "english_gloss"], sep='\t', header=None, keep_default_na=False)

hindi = pd.read_csv('original-files/tbl_all_synset-final.tsv', names=["hindi_id", "hindi_head", "hindi_synset", "hindi_gloss", "hindi_category"], sep='\t', header=None, keep_default_na=False)
```

```python
temp = map.merge(english, on='english_id')
final = temp.merge(hindi, on='hindi_id')
```

```python
final.to_csv("english-hindi.csv")
```
